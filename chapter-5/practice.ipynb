Реализуйте систему "Умный дом". В системе есть два устройства: лампа (lamp) и выключатель (switch). Вам предстоит написать сервер, который получает и обрабатывает POST запросы. Есть два типа POST запроса: \
От выключателя (switch):
```json
{
  "ip": "switch_ip",
  "id": "switch_id",
  "type": "switch",
  "action": "on" или "off" или "switch", // on - включить перечисленные лампы, off - выключить или switch - инверсировать
  "lamps": [lamp_id, lamp_id, ..., lamp_id]
}
```
От лампы (lamp):
```json
{
  "ip": "lamp_ip",
  "id": "lamp_id",
  "type": "lamp",
  "status": "on" или "off",
}
```

Если запрос приходит от выключателя - значит надо выполнить действия с лампами, перечисленными в запросе.
Если запрос приходит от лампы - значит кто-то изменил ее состояние и надо обновить информацию

Еще сервер должен каждые 3 секунды отправлять по сети запрос, чтобы получить информацию о статусе каждой известной лампы.
Лампы вовзращают json формата:
```json
{
  "ip": "lamp_ip",
  "id": "lamp_id",
  "type": "lamp",
  "status": "on" или "off",
}
```
